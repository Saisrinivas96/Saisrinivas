## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

### Ans 5

```python
ECom_Data_new.head()
```

```python
#5
a=0
b=0
c=0
d=0
for i in ECom_Data_new['discounted_price']:
    if i>600: 
        a+=i*0.25 #a = a+i*0.25
    elif i>350 and i<=600:
        b+=i*0.15
    elif i>100 and i<=350:
        c+=i*0.1
    else:
        d+=i*0.05
```

```python
print('Revenue of 25% on the orders having final price (discounted price) greater than 600 =',a)
print('Revenue of 15% on the orders having final price (discounted price) greater than 350 but less than or equal to 600 =',b)
print('Revenue of 10% on the orders having final price (discounted price) greater than 100 but less than or equal to 350 =',c)
print('Revenue of 5% on the final price (discounted price) for the orders remaining  (or) less than or equal to 100 =',d)
print('Note:There are no orders less than or equal to 100')
print('Revenue of the Ecommerse company=', a+b+c+d)
```

Ans 5) By using 'for' loop I had iterated all the list of discount_price values. After, categorised and sum up the prices based on conditiones using 'if conditional' statment. Have shown the revenue generated by ECommerce based on 25%,15%,10% and 5% conditions and sum up the values of each category for total revenue.

## Q6.Calculate the total BrandRevenue and list the top 10 Brand having maximum revenue in descending order (6 marks)
#### Hint: Total BrandRevenue is BrandRevenue which is generated after all the deductions. Also, BrandRevenue is different from the E-Commerce company's revenue.
#### Brand Revenue is the discounted price minus the Revenue (generated using the conditions given above Q5)


### Ans 6

```python
#6
#Create a function to make a list of brand revenue on each order
def price_list():
    x=[]
    for i in ECom_Data_new['discounted_price']:
        if i>600: 
            x.append(i-i*0.25)
        elif i>350 and i<=600:
            x.append(i-i*0.15)
        elif i>100 and i<=350:
            x.append(i-i*0.1)
        else:
            x.append(i-i*0.05)        
    return x
```

Above, have created a function for brand revenue of each order in the brand with for and if loop

```python
#assign the function value to new variable
y=price_list()
y
```

```python
#adding the list of brand revenue values to our data set in new coloumn called 'brand_rev'
ECom_Data_new['brand_rev']=y
ECom_Data_new.head()
```

```python
#top 10 brands
ECom_Data_new.groupby('Brand')['brand_rev'].sum().sort_values(ascending=False).head(10)
```

Ans 6)
1) Defined the brand revenu price list with a function called price_list for each order.
2) Added a new column called brand_rev in data set with list defined in price_list function.
3) Checked whether the list is added with .head() function.
4) Last step is to grouping all the list, by adding indiviual brand's revenue and sorting them for the first 10 values.
